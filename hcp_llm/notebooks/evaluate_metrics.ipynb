```python
from hcp_llm.interfaces import (
    generate_hcp_insights,
    get_summary_dict,
    PubMedScrapper,
    GenImager
)
from hcp_llm import general_settings, OpenAISettings
from hcp_llm.interfaces._Summarizer import SummarizerAI
```

Define settings & Query

```python
settings = general_settings()
settings['openai_key'] = OpenAISettings().openai_key

brand = 'COSENTYX PSA'
query = "Secukinumab psoriatic arthritis".lower()
```

Retrieve data from PubMed

```python
scrapper = PubMedScrapper(query=query)
data = scrapper.get_data()
data = data.sample(settings['sample_data'])
```

```python
for ID in data.id:
    display("#############################")
    df = data[data['id'] ==ID]
    text = df.text.unique()[0]
    display(ID)
    model = SummarizerAI(settings['openai_key'], f"""Provide the specific Abstract of the article: {text} """, settings['model'], 0)
    abstract = model.get_response()
    display(f"Abstract: {abstract}")

    model = SummarizerAI(settings['openai_key'], f"""Generate a summary (exclude Abstract) of around {len(abstract)} characters (including spaces) on the topic of the article: {text} """, settings['model'], 0)
    summary = model.get_response()
    display(f"LLM Summary: {summary}")

    prompt_evaluate = f"""I will provide an Abstract from an article & a summary generated by LLM and I need you to tell me whether the LLM is capable to properly capture same topic information:
    Abstract: {abstract}
    Summary generated by LLM: {summary}
    """
    model = SummarizerAI(settings['openai_key'], prompt_evaluate, settings['model'], 0)
    comparison = model.get_response()
    display(f"Comparison between Abstract & LLM: {comparison}")
    display("#############################")
```
